```python
import numpy as np
```

```python
class StochasticGradientDescent():
    def __init__(self,epochs):
        self.epochs=epochs
        self.coef_=None
        self.intercept_=None
        #
        self.t0=5
        #
        self.t1=50
        
    #
    def LearningRate(self,t):
        return self.t0/(t + self.t1)
        
    def fit(self,x,y):
        self.intercept_=0
        self.coef_=np.ones(x.shape[1])
        
        for i in range(self.epochs):
            for j in range(x.shape[0]):
                index=np.random.randint(0,x.shape[0])
                
                y_hat=np.dot(x[index],self.coef_) + self.intercept_
            
                intercept_slope = -2 * (y[index] - y_hat)
                # self.LearningRate(i * x.shape[0] + j)
                self.intercept_ = self.intercept_ - (self.LearningRate(i * x.shape[0] + j) * intercept_slope)
            
                coef_slope = -2 * np.dot((y[index] - y_hat),x[index])
                # self.LearningRate(i * x.shape[0] + js)
                self.coef_ = self.coef_ - (self.LearningRate(i * x.shape[0] + j) * coef_slope)
                
        print("intercept= ",self.intercept_)
        print("slope= ",self.coef_)
        
    
    def predict(self,x):
        return  np.dot(x,self.coef_) + self.intercept_
    
    #making my own Stochastic Gradient Descent class
```

```python
SGD=StochasticGradientDescent(epochs=50)

#making object of my own class
```

```python
SGD.fit(x_train,y_train)

#model training
```

```python
y_predict_2=SGD.predict(x_test)

#model testing
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict_2)

#model evaluation
```
