```python
from sklearn.datasets import load_diabetes

#importing diabetes dataset from sci-kit learn
```

```python
x,y = load_diabetes(return_X_y=True)
```

```python
print(x.shape)
print(y.shape)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

#use LinearRegression class of sci-kit learn
```

```python
lr.fit(x_train,y_train)

#model training
```

```python
y_predict=lr.predict(x_test)

#model testing
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)

#model evaluation
```

```python
print("weights: ",lr.coef_)
print("bias: ",lr.intercept_)

#getting weights and bias of trained model
```

```python
import numpy as np
```

* **Formula to find intercept**
# step: 1 intercept_slope = -2/n (Yi - Yi^)
where i from 1 to n, n=no of person/rows
# step: 2 intercept = intecept - intecpet_slope

* **Formula to find weights**
# step: 1 coef_slope = -2/n ((Y-Yi^).X)
where i from 1 to n, n=no of person/row
# step: 2 coef = coef - coef_slope

***here, X,Y,Y^ are the metrics***
# **Derrivation of above equation is available in the same repository**

```python
class BatchGradientDescent():
    def __init__(self,learning_rate,epochs):
        self.learning_rate=learning_rate
        self.epochs=epochs
        self.coef_=None
        self.intercept_=None
        
        
    def fit(self,x,y):
        self.intercept_=0
        self.coef_=np.ones(x.shape[1])
        
        for i in range(self.epochs):
            
            y_hat=np.dot(x,self.coef_) + self.intercept_
            
            intercept_slope=-2 * np.mean(y-y_hat)
            self.intercept_ = self.intercept_ - (self.learning_rate * intercept_slope)
            
            coef_slope = -2 * np.dot((y - y_hat),x)/x.shape[0]
            self.coef_ = self.coef_ - (self.learning_rate * coef_slope)
            
        print("intercept= ",self.intercept_)
        print("slope+ ",self.coef_)
        
    
    def predict(self,x):
        return  np.dot(x,self.coef_) + self.intercept_
    
    #design our own Gradient Decent class
```

```python
BGD = BatchGradientDescent(learning_rate=0.01,epochs=50)

#creating object of class
```

```python
BGD.fit(x_train,y_train)

#model training
```

```python
y_predict_2=BGD.predict(x_test)

#model testing
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict_2)

#model evaluation
```
