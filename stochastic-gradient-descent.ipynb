```python
from sklearn.datasets import load_diabetes

#importing diabetes dataset form sci-kit learn
```

```python
x,y = load_diabetes(return_X_y=True)
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)

#splitting training data and testing data
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

#importing LinearRegression class of sci-kit learn
```

```python
lr.fit(x_train,y_train)

#model training
```

```python
y_predict=lr.predict(x_test)

#model testing
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)

#model evaluation
```

```python
print("weights: ",lr.coef_)
print("bias: ",lr.intercept_)

#getting weights and bias for our dataset predicted by linearRegression class
```

```python
import numpy as np
```

**FORMULAS:-**

# **For bias**
# step 1: intercept_slope = -2(Y[index] - y^)
# step 2: intercept = intercept - intercept_slope

# **For weights**
# step 1: weight_slope = -2((Y[index]-y^).(X[index]))
# step 2: coef = coef - weight_slope

```python
class StochasticGradientDescent():
    def __init__(self,learning_rate,epochs):
        self.learning_rate=learning_rate
        self.epochs=epochs
        self.coef_=None
        self.intercept_=None
        
        
    def fit(self,x,y):
        self.intercept_=0
        self.coef_=np.ones(x.shape[1])
        
        for i in range(self.epochs):
            for j in range(x.shape[0]):
                index=np.random.randint(0,x.shape[0])
                
                y_hat=np.dot(x[index],self.coef_) + self.intercept_
            
                intercept_slope = -2 * (y[index] - y_hat)
                self.intercept_ = self.intercept_ - (self.learning_rate * intercept_slope)
            
                coef_slope = -2 * np.dot((y[index] - y_hat),x[index])
                self.coef_ = self.coef_ - (self.learning_rate * coef_slope)
                
        print("intercept= ",self.intercept_)
        print("slope= ",self.coef_)
        
    
    def predict(self,x):
        return  np.dot(x,self.coef_) + self.intercept_
    
    #making my own Stochastic Gradient Descent class
```

```python
SGD=StochasticGradientDescent(learning_rate=0.01,epochs=50)

#making object of my own class
```

```python
SGD.fit(x_train,y_train)

#model training
```

```python
y_predict_2=SGD.predict(x_test)

#model testing
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict_2)

#model evaluation
```
