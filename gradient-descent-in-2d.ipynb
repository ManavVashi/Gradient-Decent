**Equations:-**

# dl/db = -2(yi - mxi - b)
# dl/dm = -2xi(yi - mxi -b)

```python
class GradientDescent2D():
    def __init__(self,learning_rate,epochs):
        self.learninig_rate=learning_rate
        self.epochs=epochs
        self.coef_=150
        self.intercept_=2
        
    def fit(self,x,y):
        
        for i in range(self.epochs):
            slope_b = -2 * np.sum(y - (self.coef_ * x) - self.intercept_)
            slope_m = -2 * np.sum((y- (self.coef_ * x) - self.intercept_)*x)
        
            self.coef_ = self.coef_ - (self.learninig_rate*slope_m)
            self.intercept_ = self.intercept_ - (self.learninig_rate*slope_b)
    
    def predict(self,x):
        
        return self.coef_ * x + self.intercept_
    
    #creation class which can perform Gradient Decent for 2D(1 input column and 1 target column) dataset
```

# From here, we visulize our model performance with different different epochs value:-

```python
learning_rate=0.001
epochs=2
```

```python
gd2=GradientDescent2D(learning_rate,epochs)
```

```python
gd2.fit(x_train,y_train)
```

```python
y_predicted=gd2.predict(x_test)
```

```python
plt.scatter(x_train,y_train,color="blue")
plt.plot(x_train,gd2.predict(x_train),color="green")
plt.show

print("slope= ",gd2.coef_)
print("bias= ",gd2.intercept_)
```
