```python
import numpy as np
```

Creating dataset using the equation : 
# y=2x^2 +2

```python
x=np.random.rand(200,1) + np.random.randn(200,1)
```

```python
y= 2*(x**2) + 2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("x",color="red")
plt.ylabel("y",color="red")
plt.title("dataset plot",color="red")

#ploting dataset for vislulization
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)

#split data into training and testing data
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

#import LinearRegression class from sci-kit learn
```

```python
lr.fit(x_train,y_train)

#model training 
```

```python
m_actual=lr.coef_
print("slope= ",m_actual)
print("bias= ",lr.intercept_)

#getting value weight and bias derived by our model
```

* Here we doing finding line w.r.t b taking m as a constant:-
* formula is,
# b(min) at point x = dl / db = -2(yi-mxi-b) 
# Derivation is in the same repository
*where i from 1 to n
where i from 1 to n where n = no of persons or objects

```python
class GDB:
    def __init__(self,learning_rate,epochs):
        self.learning_rate=learning_rate
        self.epochs=epochs
        self.intercept_=0
        self.coef_=m_actual
        
        #here we constant the value of m/coefficient as we want to obsert the behavior of graph while changing the value of b
        #this slope value is obtained by appling linear regression 
        
    def fit_line(self,x,y):
        
        for i in range(self.epochs):
            slope_b = -2 * np.sum(y - self.coef_*x - self.intercept_)
            self.intercept_=self.intercept_-(self.learning_rate*slope_b)
        print(self.coef_,self.intercept_)
    
    #here we iterate loop for no of epochs time
    #what is epoch?? see below...
    #what is self.intercept_=self.intercept_-(self.learning_rate*slope_b) do?? see below...
    #what is learning_rate ?? see below...
    
    def predict_line(self,x):
        return (self.coef_ * x) + self.intercept_
    
    #GDB -> Gradient Descent with respect to b(intercept) where we constant the value of the slope
```

# What is epochs
 
> * epoch means traversal
> * no of epoch means no of traversal
> * How many time you traverse your dataset while model training is the no of epochs
> * example,
>> * .fit(x_train,y_train)
>> * here your model traverse your dateset only one time
>> * but if the epochs are 10 then it traverse dataset 10 times
> * if helps model to rewise it's answer about weights/slope and bias
> * also when we have less data then also it can helpful
> **We will see it in future in more detail**

# What self.intercept_=self.intercept_-(self.learning_rate*slope_b) do?? 

> * It tries to let our value of b as near as it had to be.
> * means tries to match with b obtained by OLS methode
> * it is like one like logic that perform this task
> * if values of b is too much low(let in negative) ... **self.intercept_=self.intercept_-(-x)** is up the value to positive side
> * **positive side means try to make the slope vlaue positive**
> * and if values of b is too large then... **self.intercept_=self.intercept_-(x)** is down the value to negative side
> * **negative side means try to make the slope value negative**
> * ***Int this pass - pass game it gian capability to reach to the minima in loss v/s bias graph so that what what value of b the model got minimum loss***

# What is learning rate??
> *  in this( **self.intercept_=self.intercept_-(self.learning_rate*slope_b)**) formula if we not use learning rate then, **self.intercept_=self.intercept_-*slope_b**
> * it will take big big jupms (positive to negative and negative to positive)
> * so, sometime we stuck in the loop and it visit same negative and same positive value again and again
> * Learning rate means how fast your model is learn something which is good but to avoid above problem in machine learning we try to reach to the value slowly-slowly
> * But rememner you also if you use Low learning rate then use high epochs value othervise your model training stops before reaching the actual b.
>> * it just like fuel of your cas is empty before reaching to the destination
> * in case of large data set we have to take high value of learning rate to overcome training time.
> * again it is depending upon dataset that we use high learning rate or not and It have no fixed condition,
> * This understanding comes only by the experience so don't worry right now.
> * learning rate has to in 0.something value in high and low both the cases
