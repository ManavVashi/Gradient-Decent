# Defining my own Mini-Batch Gradient Descent class

```python
class MiniBatchGradientDescent():
    def __init__(self,learning_rate,epochs,batch_size):
        self.learning_rate=learning_rate
        self.epochs=epochs
        self.batch_size=batch_size
        self.coef_=None
        self.intercept_=None
        
    def fit(self,x,y):
        self.coef_=np.ones(x.shape[1])
        self.intercept_=0
        
        for i in range(self.epochs):
            for j in range(int(x.shape[0] / self.batch_size)):
                
                index = np.random.randint(0,x.shape[0],self.batch_size)
                
                y_hat = np.dot(x[index],self.coef_) + self.intercept_
                
                # for intecept
                intercept_slope = -2 * np.mean(y[index]-y_hat)
                self.intercept_ = self.intercept_ - (self.learning_rate * intercept_slope)
                
                coef_slope = -2 * np.dot((y[index] - y_hat),x[index])
                self.coef_ = self.coef_ - (self.learning_rate * coef_slope)
                
    
    def predict(self,x):
        return np.dot(x,self.coef_) + self.intercept_
```

```python
MBGD1 =  MiniBatchGradientDescent(learning_rate=0.001,epochs=50,batch_size=10)
```

```python
MBGD1.fit(x_train,y_train)
```

```python
print("bias: ",MBGD1.intercept_)
print("weights: ",MBGD1.coef_)
```

```python
y_predict=MBGD1.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```
