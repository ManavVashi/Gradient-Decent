# * **We did't have direct class for performing Mini-Batch Gradient Descent in sci-kit learn**
# * **But we can perform it using SGDRegressor class of sci-kit learn**

```python
from sklearn.datasets import load_diabetes
x,y = load_diabetes(return_X_y=True)

#importing and loading diabetes dataset form sci-kit learn
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)

#splitting training data and testing data
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# here is the code,

```python
from sklearn.linear_model import SGDRegressor
MBGD = SGDRegressor(learning_rate='constant',eta0=0.01,max_iter=300)
```

# logic,
# SGDRegressor.partial_fit(x,y) iterate only one time

```python
batch_size=20

index=np.random.randint(0,x_train.shape[0],batch_size)
```

```python
print("weights: ",MBGD.coef_)
print("bias: ",MBGD.intercept_)
```

```python
y_predict=MBGD.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```
