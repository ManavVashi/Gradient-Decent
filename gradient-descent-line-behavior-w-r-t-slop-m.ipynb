# Note:

# In that I written theory and perameter description.

* Here we doing finding line w.r.t m taking b as a constant:-
# * formula is dl / db = -2xi(yi-mxi-b) 
# * derivaion is in the same repository
*where i from 1 to n
where i from 1 to n where n = no of persons or objects

```python
class GDM:
    def __init__(self,learning_rate,epochs):
        self.learning_rate=learning_rate
        self.epochs=epochs
        self.coef_=100
        self.intercept_=b_actual
        
    def fit_line(self,x,y):
        for i in range(self.epochs):
            slope_m = -2 * np.sum((y - self.coef_*x - self.intercept_)*x)
            self.coef_=self.coef_-(self.learning_rate*slope_m)
    
    def predict_line(self,x):
        return (self.coef_ * x) + self.intercept_
    
    #creating calss of GDM
    # GDM -> Gredient Descent with respect to weights
```

# From here, we visulize our model performance with different different epochs value:-

# 1)

```python
learning_rate=0.001
epochs=2
```

```python
object1=GDM(learning_rate,epochs)
```

```python
object1.fit_line(x_train,y_train)
```

```python
y_predicted=object1.predict_line(x_test)
```

```python
plt.scatter(x_train,y_train,color="blue")
plt.plot(x_train,object1.predict_line(x_train),color="green")
plt.show

print("slope= ",object1.coef_)
print("bias= ",object1.intercept_)
```
